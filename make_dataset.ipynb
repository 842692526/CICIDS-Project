```python
import numpy as np
import pandas as pd
import time
import csv
```

```python
Monday = pd.read_csv('Tuesday.csv')
feature=['td','sa','da','sp','dp','pr','ipkt','ibyt','opkt','obyt','Label']
M = Monday[feature]
M = M.to_numpy()

M

file_datas = []

for datas in M:
    file_data = {}
    td,sa,da,sp,dp,pr,ipkt,ibyt,opkt,obyt,Label = datas
    file_data['td'] = td
    sa = sa.split('.')
    sa = [int(x) for x in sa]
    file_data['sa_1'] = sa[0]
    file_data['sa_2'] = sa[1]
    file_data['sa_3'] = sa[2]
    file_data['sa_4'] = sa[3]

    da = da.split('.')
    da = [int(x) for x in da]
    file_data['da_1'] = da[0]
    file_data['da_2'] = da[1]
    file_data['da_3'] = da[2]
    file_data['da_4'] = da[3]

    file_data['sp'] = sp
    file_data['dp'] = dp
    file_data['pr'] = pr
    file_data['ipkt'] = ipkt
    file_data['ibyt'] = ibyt
    file_data['opkt'] = opkt
    file_data['obyt'] = obyt
    file_data['Label'] = Label

    file_datas.append(file_data)
    


data = pd.DataFrame(file_datas)
fileName = 'Tue.csv'
feature=['td','sa_1','sa_2','sa_3','sa_4','da_1','da_2','da_3','da_4','sp','dp','pr','ipkt','ibyt','opkt','obyt','Label']
data.to_csv(fileName, header=feature, index=False, mode='a+', encoding='utf-8')
```
